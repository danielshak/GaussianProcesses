```python
import pandas as pd
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
import pyKriging
from pyKriging.krige import kriging  
from pyKriging.samplingplan import samplingplan
from sklearn.metrics import mean_squared_error

%matplotlib inline
```

```python
vsp_data = pd.read_csv('../Data/vsp_input.tsv',sep='\t')
chrm1_data = vsp_data[vsp_data['AvgChrs']<=1]
chrm1_data = chrm1_data[chrm1_data['WSU_Old']!=17]
chrm1_data.head()
```

```python
#Subset data into different partitions to test how much data the GP needs
data_percent = 0.5

msk = np.random.rand(len(chrm1_data)) < data_percent
train_chrm1 = chrm1_data[msk]
test_chrm1 = chrm1_data[~msk]
```

```python
x=train_chrm1[['AvgChrs','AT2_18M_F10']].values
y=train_chrm1[['WSU_Old']].values.squeeze()
```

```python
k = kriging(x, y, name='simple')
k.train()
```

```python
k.plot()
```

```python
x_test = test_chrm1[['AvgChrs','AT2_18M_F10']].values
y_test = test_chrm1[['WSU_Old']].values.squeeze()
```

```python
y_pred=[]
for position in x_test:
    y_pred.append(k.predict(position))
```

```python
mean_squared_error(y_test,y_pred)
```

```python
pearsonr = scipy.stats.pearsonr(y_test,y_pred)
print(pearsonr)
```

```python
plt.scatter(y_test,y_pred)
```

```python
results=np.array([y_test,y_pred]).T
```

```python
results[results[:,0].argsort()]
```
